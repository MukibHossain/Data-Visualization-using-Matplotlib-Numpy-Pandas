```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_dict = {
    "Sales" : [10, 20, 30, 40],
    "Profits" : [2, 4, 3, 1],
    "Date" : ["2022-01-01", "2022-02-01", "2022-03-01", "2022-04-01"]
}

df= pd.DataFrame(
    data_dict,
    columns = ["Sales", "Profits"],
    index = data_dict["Date"]
)
df
```

```python
fig, ax = plt.subplots()
ax.plot(df.index, df["Sales"], ls = "--")
ax.plot(df.index, df["Profits"], ls = "--")
ax.set_title("Product Sales and Profits")
```

```python
data_mukib = {
    "Month":["Janu" , "Feb", "March", "April", "May", "June", "July"],
    "Income":[1000, 2000, 1500, 3000, 2000, 1300, 1500],
    "Spending":[500, 300, 200, 700, 600, 900, 750]
}
df = pd.DataFrame(
        data_mukib,
    columns =["Income", "Spending"],
    index= data_mukib["Month"]
)
```
