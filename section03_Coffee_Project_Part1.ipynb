```python
top10_producers = coffee.sum().sort_values(ascending = False).iloc[:10]

top10_producers.head()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.bar(x = top10_producers.index, height=top10_producers/1000000)

positions = list(range(10))

labels = ["BR", "VN", "CO", "ID", "ET", "IN", "MX", "GT", "HN", "UG"]

ax.set_title("Top coffee by nations")
ax.set_ylabel("Productions (Millions 60 kg bags)", fontsize = 10)
ax.set_xlabel("Country Name", fontsize = 10)

ax.set_xticks(positions)
ax.set_xticklabels(labels)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

# Top 5 over time

* Create a line chart with the top 5 nations over time. Make sure to include a legend! Consider converting the index to datetime64[ns].

```python
top5 = coffee.loc[:, ["Brazil", "Viet Nam", "Colombia", "Indonesia", "Ethiopia"]]
top5.index = top5.index.astype("datetime64[ns]")

top5.head()
```

```python
fig, ax = plt.subplots()

ax.plot(top5.index, coffee["Brazil"] / 1000, label="Brazil")
ax.plot(top5.index, coffee["Viet Nam"] / 1000, label="Viet Nam")
ax.plot(top5.index, coffee["Colombia"] / 1000, label="Colombia")
ax.plot(top5.index, coffee["Indonesia"] / 1000, label="Indonesia")
ax.plot(top5.index, coffee["Ethiopia"] / 1000, label="Ethiopia")


ax.set_title("Top coffee by nations")
ax.set_ylabel("Productions (Millions 60 kg bags)", fontsize = 10)
ax.set_xlabel("Country Name", fontsize = 10)

ax.set_xticks(positions)
ax.set_xticklabels(labels)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.legend()
```
